```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
plt.figure(figsize=(10, 8))
sns.barplot(data=df3, x="modzcta_first",y = "NUM_PEOP_TEST")
plt.xlabel("Zip Code", fontsize=10)
plt.ylabel("No. of test", fontsize=10)
plt.title('Covid test by Zip code', fontsize=20)
plt.legend()
sns.set(font_scale = .5)
plt.show()
```

```
sns.lineplot(data=df3, x="label",y = "NUM_PEOP_POS",color = 'violet', linewidth = 2.5)
plt.xlabel("Zip Code", fontsize=10)
plt.ylabel("No. of test", fontsize=10)
plt.title('Covid Prositive by Zip code', fontsize=20)
plt.show()
```
